```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Translation of Line
X=np.arange(1,6)
Y=5-X
X2=X+2
Y2=Y+3
```

```python
fig=plt.figure(figsize=(10,8))
plt.plot(X,Y,label="Original Line",)
plt.plot(X2,Y2,label="Translated Line (translated by (2,3)")
plt.legend()
plt.title("Translation Of Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```

```python
#Translation of Curve
X3=np.arange(0,4,0.1)
Y3=(16-(X3**2))**(1/2)
X4=X3+1
Y4=Y3+2
```

```python
fig=plt.figure(figsize=(10,8))
plt.plot(X3,Y3,label="Original Curve",)
plt.plot(X4,Y4,label="Translated Curve (translated by (1,2))")
plt.legend()
plt.title("Translation Of Curve")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```
